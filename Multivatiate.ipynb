```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('/content/Untitled spreadsheet (4).xlsx')

X = data[['X1', 'X2']]
y = data['Y']

X['bias'] = 1

X = X.to_numpy()
y = y.to_numpy()

theta = np.zeros(X.shape[1])

learning_rate = 0.01
num_iterations = 1000

for _ in range(num_iterations):

    predictions = np.dot(X, theta)

    errors = predictions - y

    gradients = np.dot(X.T, errors) / len(y)

    theta -= learning_rate * gradients

print("Learned Coefficients:", theta)

plt.scatter(X[:, 0], y, label='Actual data', color='blue')
plt.scatter(X[:, 0], predictions, label='Predicted data', color='red')
plt.plot(X[:, 0], predictions, label='Regression line', color='green')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Scatter Plot with Regression Line')
plt.legend()
plt.show()

```
