#LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,5,2,6]).reshape((-1,1))
y=np.array([3,5,9,6,4])
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='orange',marker='+')

print("MSE:",mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random
```

```python
random.seed(42)
num_samples = 30
x = np.array([random.uniform(1, 30) for _ in range(num_samples)]).reshape((-1, 1))
y = np.array([2 * val + random.uniform(-5, 5) for val in x])

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)

mse = mean_squared_error(y, pred)
plt.plot(x, pred, color='black', marker='+')

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
